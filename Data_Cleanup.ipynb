```python
import pandas as pd
import numpy as np
import csv

df = pd.read_json("cardinfo.json", orient='split')
df.head()
```

```python
df = df.drop(['id', 'frameType', 'card_sets', 'card_images', 'card_prices', 'linkmarkers'], axis=1)
df = df[df['type'] != "Skill Card"]
df['name'] = df['name'].astype(pd.StringDtype())
df['type'] = df['type'].astype(pd.StringDtype())
df['desc'] = df['desc'].astype(pd.StringDtype())
df['race'] = df['race'].astype(pd.StringDtype())
df['archetype'] = df['archetype'].astype(pd.StringDtype())
df['attribute'] = df['attribute'].astype(pd.StringDtype())
```

```python
def search(regex: str, df, case=False):
    """https://stackoverflow.com/questions/26640129/search-for-string-in-all-pandas-dataframe-columns-and-filter"""
    """Search all the text columns of `df`, return rows with any matches."""
    textlikes = df.select_dtypes(include=[object, "string"])
    return df[
        textlikes.apply(
            lambda column: column.str.contains(regex, regex=True, case=case, na=False)
        ).any(axis=1)
    ]

search("Pot Of Greed", df).head()
```

```python
def on_banlist(banlist_info):
    results = []
    for i in banlist_info:
        if pd.isnull(i):
            results.append(0)
        else:
            if 'ban_tcg' in i:
                results.append(1)
            else:
                results.append(0)
    return results
        
df['meta'] = on_banlist(df['banlist_info'])
df = df.drop(['banlist_info'], axis=1)
search("qliphort scout", df).head()
```

```python
def create_sentence(df):
    results = []
    for index, row in df.iterrows():
        archetype = ""
        if not pd.isnull(row["archetype"]):
            archetype = "{} is a part of the {} archetype.".format(row['name'], row['archetype'])
        
        desc = row['desc']
        if row['type'] == "Normal Monster":
            desc = ""  
        elif row['type'] == "Pendulum Normal Monster":
            sep = '[ Flavor Text ]'
            desc = row['desc'].split(sep, 1)[0]
        
        if row['type'] == "Spell Card" or row['type'] == "Trap Card":
            results.append("{} is a {} {}. {} {}".format(row['name'], row['race'], row['type'], archetype, desc))
        elif not pd.isnull(row['scale']):
            results.append("{} is a level {} {} {} {} with {} attack, {} defense, and a pendulum scale of {}. {} {}".format(
                row['name'], int(row['level']), row['attribute'], row['race'], row['type'], int(row['atk']), int(row['def']), int(row['scale']), archetype, desc))
        elif not pd.isnull(row['linkval']):
            results.append("{} is a link {} {} {} {} with {} attack. {} {}".format(
                row['name'], int(row['linkval']), row['attribute'], row['race'], row['type'], int(row['atk']), archetype, desc))
        else:
            if pd.isnull(row['level']): print(row['name'])
            results.append("{} is a level {} {} {} {} with {} attack, and {} defense. {} {}".format(
                row['name'], int(row['level']), row['attribute'], row['race'], row['type'], int(row['atk']), int(row['def']), archetype, desc))
    return results

df['info_sentence'] = create_sentence(df)
df = df.drop(['name', 'type', 'race', 'archetype', 'atk', 'def', 'level', 'attribute', 'scale', 'linkval'], axis = 1)
print(df.iloc[1]['info_sentence'])
#df.head()
```
