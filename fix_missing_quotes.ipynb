```python
with open('card_info_clean_with_meta.csv', 'r') as bad_file:
    # read a list of lines into data
    bad_data = bad_file.readlines()

for i, line in enumerate(bad_data):
    if ".,0" in line:
        #print(line.split('.,0', 1)[0])
        bad_data[i] = "\"" + line.split('.,0')[0] + ".\",0"
        print(bad_data[i])
    elif ".,1" in line:
        bad_data[i] = "\"" + line.split(".,1")[0] + ".\",1"

# and write everything back
with open('fixed.csv', 'w') as file:
    file.writelines(bad_data)
```
